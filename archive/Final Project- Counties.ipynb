# Fighting Fire in California

### Research question: 
What external factors influence average fire salary?

### Data:
https://transparentcalifornia.com/agencies/salaries/#counties
> This link contains information about salaries in each county of California.
> Most datasets have data from 2016 except one county so we eliminated that one.

```python
import numpy as np
import pandas as pd
import geopandas as gpd
import os
import glob
```

```python
# put all csv names in array
os.chdir('data/counties') 
files = glob.glob('*-2016.csv')

#for testing
test_files = ['los-angeles-county-2016.csv',
 'san-mateo-county-2016.csv',
 'san-diego-county-2016.csv']

len(files)
```

```python
#combine all csvs into three dataframes
#first one, 0-19
combined_one = pd.concat([pd.read_csv(f) for f in files[0:20]])
combined_one.to_csv("combined_one.csv", index=False, encoding='utf-8-sig')

#second one, 20-40
combined_two = pd.concat([pd.read_csv(f) for f in files[20:40]])
combined_two.to_csv("combined_two.csv", index=False, encoding='utf-8-sig')

#third one, 40-end
combined_three = pd.concat([pd.read_csv(f) for f in files[40:]])
combined_three.to_csv("combined_three.csv", index=False, encoding='utf-8-sig')

combined_one
```

```python
#dropping useless columns
combined_one = combined_one.drop(columns=['Base Pay', 'Overtime Pay', 'Other Pay', 'Benefits', 'Total Pay & Benefits', 'Year', 'Notes', 'Status'])
combined_two = combined_two.drop(columns=['Base Pay', 'Overtime Pay', 'Other Pay', 'Benefits', 'Total Pay & Benefits', 'Year', 'Notes', 'Status'])
combined_three = combined_three.drop(columns=['Base Pay', 'Overtime Pay', 'Other Pay', 'Benefits', 'Total Pay & Benefits', 'Year', 'Notes', 'Status'])

combined_one
```

```python
# all job titles with "fire"
fire_one = combined_one[combined_one['Job Title'].str.lower().str.contains('fire')]
fire_two = combined_two[combined_two['Job Title'].str.lower().str.contains('fire')]
fire_three = combined_three[combined_three['Job Title'].str.lower().str.contains('fire')]

fire_one
```
